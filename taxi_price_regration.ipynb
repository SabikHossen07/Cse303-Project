### Weather Condition

```python

df['Weather'].value_counts().plot(kind='bar', color='purple', edgecolor='black')
plt.title("Trips by Weather Condition")
plt.xlabel("Weather Condition")
plt.ylabel("Number of Trips")
plt.show()
```

### Passenger Count

```python

df['Passenger_Count'].value_counts().plot(kind='bar', color='blue', edgecolor='black')
plt.title("Trips by Passenger Count")
plt.xlabel("Passenger Count")
plt.ylabel("Number of Trips")
plt.show()
```

### Percentage of Trips by Weather Condition

```python
colors = plt.cm.Pastel1.colors 
df['Weather'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    colors=colors
)
plt.title("Percentage of Trips by Weather Condition")
plt.ylabel("")
plt.show()
```

### Average Fare by Trip Duration

```python
avg_fare = df.groupby('Trip_Duration_Minutes')['Trip_Price'].mean()
plt.figure(figsize=(8,5))
plt.scatter(avg_fare.index, avg_fare.values, color='teal')
plt.title("Average Fare by Trip Duration", fontsize=14)
plt.xlabel("Trip Duration (Minutes)")
plt.ylabel("Average Fare (USD)")
plt.show()
```
